```python
import pandas as pd
import plotly.express as px
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/suparuek2405/DADS5001_Tool/refs/heads/main/listings.csv')
df.head()
```

```python
# prompt: create new dataframe and remove price outlier

# Calculate the IQR for the 'price' column
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create a new DataFrame without outliers
df_no_outliers = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
```

```python
fig = px.box(df_no_outliers, x='room_type', y='price')
fig.show()
```

```python
fig = px.scatter(df_no_outliers, x='minimum_nights', y='price')
fig.show()
```

```python
# prompt: visualize using latitude and longtitude and add color as price

fig = px.scatter_mapbox(df_no_outliers, lat="latitude", lon="longitude", color="price", zoom=10)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
